### Medical Info Class

```python
class medical_info:
    def __init__(self, csv):
        self.df_insurance = pd.read_csv(csv)
        # grouping the data for futher analysis
        self.grp_sex = self.df_insurance.groupby("sex")
        self.grp_region = self.df_insurance.groupby("region")
        self.grp_smoker = self.df_insurance.groupby("smoker")
        self.grp_region_smoker = self.df_insurance.groupby(["region", "smoker"])
        
    def counts(self):
        self.males = self.df_insurance["sex"].loc[df_insurance["sex"] == "male"].count()
        self.females = self.df_insurance["sex"].loc[df_insurance["sex"] == "female"].count()
        self.smokers = df_insurance.groupby("smoker").get_group("yes")["smoker"].count()
        self.nonsmokers = df_insurance.groupby("smoker").get_group("no")["smoker"].count()
        self.count_sex = self.grp_sex['age'].count()
        self.count_region = self.grp_region['age'].count()
        self.count_smokers = self.grp_smoker['age'].count()
        return (f"Number of Males and Females: \n {self.count_sex} \n\n The number in each region:\n{self.count_region}" +
        f"\n\n The number of smokers and non-smokers:\n{self.count_smokers}")
        
    def averages(self):
        self.mean = round(self.df_insurance[["age", "bmi", "children", "charges"]].mean(), 2)
        self.age_mean, self.bmi_mean, self.children_mean, self.charges_mean = self.mean
        # self.age_mean = self.df_insurance["age"].mean()
        # self.region_mean = self.df_insurance.groupby("region").mean()
        # self.bmi_mean = self.df_insurance["bmi"].mean()
        # self.sex_mean = self.df_insurance.groupby("sex").mean()
        # return (f"The average age in the data is: {round(self.age_mean,1)} \n\nThe averages per region are: \n {self.region_mean}" + 
        # f"\n\nthe averages per sex are: \n {self.sex_mean}")
        # return f"the average bmi is {round(self.bmi_mean, 2)}"
        return str.format("The average age is {}, average bmi is {}, average num of children is {}, average insurance charges are ${}", self.age_mean, self.bmi_mean, self.children_mean, self.charges_mean)
    
    def average_charges(self):
        self.mean_charges = self.df_insurance["charges"].mean()
        return f"The average medical ins charges of the data are: ${round(self.mean_charges, 2)}"
    
    def __repr__(self):
        return "Here are basic statistics on the dataset: \n" + str(df_insurance.describe())
    
ins1 = medical_info("insurance.csv")
print(ins1)
print("\n")
print(ins1.counts())
print("\n") 
print(ins1.averages())
print("\n\n")
print(ins1.average_charges())
print("\n\n")
print(ins1.mean)
ins1.grp_sex["charges"].mean()
```
